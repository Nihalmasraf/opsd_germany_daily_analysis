```python
import pandas as pd

# Load the dataset from the given URL
url = "https://raw.githubusercontent.com/salemprakash/EDA/refs/heads/main/Data/indian_food.csv"
df = pd.read_csv(url)

# Display the first few rows of the dataset
print(df.head())

# Get basic information about the dataset
print(df.info())

# Summary statistics for numerical columns
print(df.describe())

# Checking for missing values
print(df.isnull().sum())


# Mean, Median, Mode for numerical columns
mean_prep_time = df['prep_time'].mean()
median_prep_time = df['prep_time'].median()
mode_prep_time = df['prep_time'].mode()

print(f"Mean prep time: {mean_prep_time}")
print(f"Median prep time: {median_prep_time}")
print(f"Mode prep time: {mode_prep_time[0]}")

# Standard deviation, Variance for prep_time
std_prep_time = df['prep_time'].std()
var_prep_time = df['prep_time'].var()

print(f"Standard deviation of prep time: {std_prep_time}")
print(f"Variance of prep time: {var_prep_time}")

import matplotlib.pyplot as plt

# Histogram of preparation time
plt.figure(figsize=(10, 6))
plt.hist(df['prep_time'].dropna(), bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Preparation Time')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Bar plot for the flavor profile
df['flavor_profile'].value_counts().plot(kind='bar', color='green', figsize=(10, 6))
plt.title('Flavor Profile Count')
plt.xlabel('Flavor Profile')
plt.ylabel('Count')
plt.show()

# Pie plot for course distribution
df['course'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Course Distribution')
plt.ylabel('')
plt.show()
```
